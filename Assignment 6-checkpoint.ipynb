```python
#240

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix or not matrix[0]:
            return False

        rows = len(matrix)
        cols = len(matrix[0])
        row = 0
        col = cols - 1

        while row < rows and col >= 0:
            if matrix[row][col] == target:
                return True
            elif matrix[row][col] > target:
                col -= 1
            else:
                row += 1

        return False
```

```python
#885

class Solution:
    def spiralMatrixIII(self, rows: int, cols: int, rStart: int, cStart: int) -> List[List[int]]:
        result = []
        steps = 0  
        r, c = rStart, cStart
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  
        dir_index = 0  
        while len(result) < rows * cols:
            if dir_index == 0 or dir_index == 2:
                steps += 1
            for _ in range(steps):
                if 0 <= r < rows and 0 <= c < cols:
                    result.append([r, c])
                dr, dc = directions[dir_index]
                r += dr
                c += dc
            dir_index = (dir_index + 1) % 4
        return result
```

```python
#1463

class Solution:
    def cherryPickup(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        dp = [[[0]*n for _ in range(n)] for _ in range(m)]
        for c1 in range(n):
            for c2 in range(n):
                dp[m-1][c1][c2] = grid[m-1][c1] + (grid[m-1][c2] if c2 != c1 else 0)
        for row in range(m-2, -1, -1):
            for c1 in range(n):
                for c2 in range(n):
                    max_cherries = 0
                    for nc1 in (c1-1, c1, c1+1):
                        for nc2 in (c2-1, c2, c2+1):
                            if 0 <= nc1 < n and 0 <= nc2 < n:
                                max_cherries = max(max_cherries, dp[row+1][nc1][nc2])

                    current = grid[row][c1] + (grid[row][c2] if c2 != c1 else 0)
                    dp[row][c1][c2] = current + max_cherries
        return dp[0][0][n-1]
```

```python
#1861

class Solution:
    def rotateTheBox(self, box: List[List[str]]) -> List[List[str]]:
        m, n = len(box), len(box[0])
        rotated = [[None] * m for _ in range(n)]
        for i in range(m):
            for j in range(n):
                rotated[j][m - 1 - i] = box[i][j]
        for col in range(m):
            empty_positions = deque()
            for row in range(n - 1, -1, -1):
                if rotated[row][col] == '*':
                    empty_positions.clear()
                elif rotated[row][col] == '.':
                    empty_positions.append(row)
                elif empty_positions:
                    drop_row = empty_positions.popleft()
                    rotated[drop_row][col] = '#'
                    rotated[row][col] = '.'
                    empty_positions.append(row)
        return rotated
```

```python
#1975

class Solution:
    def maxMatrixSum(self, matrix: List[List[int]]) -> int:
        total = 0
        neg_count = 0
        min_abs = float('inf')

        for row in matrix:
            for val in row:
                abs_val = abs(val)
                total += abs_val
                if val < 0:
                    neg_count += 1
                min_abs = min(min_abs, abs_val)

        if neg_count % 2 != 0:
            total -= 2 * min_abs

        return total
```
