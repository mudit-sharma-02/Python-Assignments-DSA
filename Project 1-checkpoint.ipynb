# Question 3
Create 2 new columns based on Player name. First column will have first name and second column will have last name. Eg: for the player Shikhar Dhawan, Shikhar will be the first name and Dhawan will be the last name.

# Question 4
Create a new column (Cleaned_Highest_score) based on Highest score variable. Remove the Asterik(*) mark and convert the data type into INT.

```python
## To remove the Asterik,we have to use Replace function
ipl['Cleaned_Highest_score'] = ipl['Highest Score'].str.replace('*','')
ipl['Cleaned_Highest_score']
```

```python
## Now we see NaN in upper Output - so to remove that we have to use .astype(str) funtion
ipl['Cleaned_Highest_score'] = ipl['Highest Score'].astype(str).str.replace('*','')
ipl['Cleaned_Highest_score']
```

```python
## Now to convert colum into integers we use .astype(int) function
ipl['Cleaned_Highest_score'] = ipl['Cleaned_Highest_score'].astype(int)
ipl['Cleaned_Highest_score']
```

```python
##Final Result
print(ipl[['Player','Highest Score','Cleaned_Highest_score']])
```

```python
##Extra
ipl['Cleaned_Highest_score'] = ipl['Highest Score'].astype(str).str.replace('*','',regex = False)
##regex(regular expression) = it's simply removes the special characters and helps to prevent error or any unexpected behavior
ipl['Cleaned_Highest_score']
```

# Question 17
Create a new column to show the percentage of total runs scored in 4s and 6s. Then print the top 5 players with maximum percentage.

```python
## At first we'll calculate runs from boundaries
ipl['Runs_in_Boundaries'] = (ipl['4s']*4) + (ipl['6s']*6)
ipl['Runs_in_Boundaries']
```

```python
## Now we have to calculate percentage
ipl['Boundary_Percentage'] = (ipl['Runs_in_Boundaries'] / ipl['Runs'])*100
ipl['Boundary_Percentage']
```

```python
Top5_percentage = ipl.sort_values(by = 'Boundary_Percentage',ascending = False).head(5)[['Player','Runs','Runs_in_Boundaries','Boundary_Percentage']]
Top5_percentage
```

# Question 19
Create visualization of top 10 players with highest number of sixes.

```python
Top_10 = ipl.sort_values(by = '6s',ascending = False).head(10)[['Player','6s']]
Top_10
```

```python
plt.figure(figsize = (15,7))
sns.barplot(x = 'Player',y= '6s',data = Top_10,palette='plasma')

plt.title("Top 10 Players with Highest Number of Sixes", fontsize=16)
plt.xlabel("Players")
plt.ylabel("Number of Sixes")
plt.xticks(rotation=45 , ha="right")
plt.show()
```

```python
## Method 2
plt.figure(figsize = (15,7))
plt.bar(Top_10['Player'],Top_10['6s'],color = 'orange',edgecolor = 'red')

plt.title('Top 10 players with the highest no.s of sixes',fontsize = 15)
plt.xlabel('6s')
plt.ylabel('Players')

plt.xticks(rotation=90)
plt.show()
```

# Question 20
Scatter plot of runs scored by a player v/s balls faced in a season. Then find the relationship between these 2 variables.

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='Balls faced',y='Runs',data = ipl,color = 'blue',s=50)

plt.title('Scatter plot of runs scored by a player v/s balls faced',fontsize = 15)
plt.xlabel('Ball Faced')
plt.ylabel('Runs')

plt.xticks(range(0,540,20),rotation = 90)
plt.yticks(range(0,750,30))
plt.show()
```

```python
## Now we have to check corelation
Correlation = ipl['Runs'].corr(ipl['Balls faced'])
print(Correlation)   ##So the output shows the Positive Relationship
```
